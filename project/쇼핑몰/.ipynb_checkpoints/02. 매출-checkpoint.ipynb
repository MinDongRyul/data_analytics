### 해당 기간의 매출
- 전체 매출
- 국가별 매출
- 월별 매출
- 요일별 매출 
- 시간별 매출

```python
# 전체 매출

total_revenue = retail['CheckoutPrice'].sum()
total_revenue
```

```python
# 국가별 매출

rev_by_contries = retail.groupby('Country')[['CheckoutPrice']].sum().sort_values('CheckoutPrice',ascending=False)
rev_by_contries
```

```python
ax = rev_by_contries.plot.bar(figsize=(20,10))
ax.set_xlabel('Country', fontsize=11)
ax.set_ylabel('Revenue', fontsize=11)
ax.set_title('Revenue by Country', fontsize=13)
```

```python
rev_by_contries / total_revenue
```
