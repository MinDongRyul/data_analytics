## 데이터 전처리

### date,time 컬럼 변환한 후 하나의 컬럼 생성

```python
lakers.dtypes
```

```python
# date 컬럼 str타입으로 변경
lakers['date'] = lakers['date'].astype('str')
# datetime = date + ' ' + time
lakers['datetime'] = lakers['date'] + " " + lakers['time']
```

```python
lakers.dtypes
```

```python
lakers.head()
```

```python
lakers['date'] = pd.to_datetime(lakers['date'], format='%Y%m%d')
lakers['datetime'] = pd.to_datetime(lakers['datetime'], format='%Y%m%d %H:%M:%S')
```

```python
lakers.head()
```

```python
lakers.groupby('game_type').sum()
```

```python
# 날짜별로 홈경기인지 원정경기인지?

lakers[['date','game_type']].groupby(['date','game_type']).sum() # 여기서의 sum() 그냥 데이터를 끌어내기 위한 것
```

```python
df1 = lakers.drop_duplicates(['date'],keep='first')
df1
```

```python
f, ax = plt.subplots(1,2, figsize=(18,8))

# 원 차트
df1['game_type'].value_counts().plot.pie(autopct='%1.1f%%', ax = ax[0], shadow = True, explode=[0,0.1])
ax[0].set_title('홈경기 vs 원정경기', size=15)
ax[0].set_ylabel('')

# count plt
sns.countplot('game_type',data=df1, ax=ax[1])
ax[1].set_title('홈경기 vs 원정경기', size=15)

plt.show()
```

## 이번 시즌 경기 결과는?

### 1) 경기 결과 중에서 ~~

```python
# 날짜별 경기 결과에서 result 컬럼의 value가 made인 points만 더하면 총 점수

# lakers['result'] == 'made'

lakers['result'].isin(['made'])
```

#### 2) made경기에 총 점수 구하기

```python
df3 = lakers.query('result == "made"').groupby(['date','team'])[['points']].sum().reset_index()
df3
```

#### 3) LA레이커스와 상대편으로 경기 결과 나누기

```python
lal = df3[df3['team'] == 'LAL']
opp = df3.query('team != "LAL"')
```

```python
opp.head(3)
```

```python
lal.head(3)
```

#### 4) 경기 결과에 대해서 merge

```python
df4 = pd.merge(lal,opp, on='date')
df4
```

```python
# 승리 횟수
df4[df4['points_x'] > df4['points_y']].shape[0]

win = df4[df4['points_x'] > df4['points_y']]
los = df4[df4['points_x'] < df4['points_y']]
win
```

```python
# LA레이커스가 이긴 날과 진 날을 알아본 후 출력

# 08-09 시즌에 LA레이커스는 총 {}회의 경기 중 {}번 승리했다.

print('08-09 시즌에 LA레이커스는 총 {}회의 경기 중 {}번 승리했다.'.format(df4.shape[0],df4[df4['points_x'] > df4['points_y']].shape[0]))
```

```python
plt.figure(figsize=(16,5))

plt.title('08 - 09 Season : LA Lakers')

plt.plot(lal['date'],lal['points'], '-', alpha=1, color='orange', label='LA Lakers')

plt.plot(win['date'], win['points_x'], 'o', alpha=0.6, color='red')
plt.plot(los['date'], los['points_x'], 'o', alpha=0.6, color='blue')

plt.plot(opp['date'], opp['points'], '--', alpha=0.6, color='green', label='Opposite team')

plt.legend(loc='upper left', fontsize=13)
plt.xticks(rotation=45)

plt.show()
```
