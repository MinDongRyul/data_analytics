#### ,가 붙은 values를 숫자의 형태로 변경

```python
import numpy as np

data = {
    'name':['홍길동','김수정','이민정'],
    'age':['25',np.nan,'45']
}
df = pd.DataFrame(data)
df
```

```python
df['age_1'] = df['age'].astype('int64')
```

- astype() : values가 NaN을 포함하고 있으면 에러 발생

```python
# pd.to_numeric(변경하고자할 컬럼)

df['age_1'] = pd.to_numeric(df['age'])
df
```

# 2022년 5월 기준

```python
df_man2 = pd.read_excel("../../data/dataset/202205_202205_연령별인구현황_월간.xlsx",skiprows=3, usecols='B,E:Y', index_col='행정기관',thousands=',')
```

```python
df_woman2 = pd.read_excel("../../data/dataset/202205_202205_연령별인구현황_월간.xlsx",skiprows=3, usecols='B,AB:AV', index_col='행정기관',thousands=',')
```

```python
df_man2.head(2)
```

```python
df_woman2.head(2)
```

```python
df_woman2.columns = df_man2.columns
```

```python
# str을 숫자형으로 바꿔줌 ',' -> '' : , 부분을 없애기

df_woman2 = df_woman2.apply(lambda x:x.str.replace(',','').apply('int64'))
```

```python
df_woman2.head(2)
```

```python
df_woman2.info()
```

# 시각화

```python
plt.figure(figsize=(10,7))

plt.barh(y=df_man2.columns,width = -df_man2.iloc[0] // 10000)
plt.barh(y=df_woman2.columns,width = df_woman2.iloc[0] // 10000)
plt.title('2022년 5월 기준 대한민국 인구 피라미드')
plt.savefig('2022_인구피라미드.png')
plt.show()
```
