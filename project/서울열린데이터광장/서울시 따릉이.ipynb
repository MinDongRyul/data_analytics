### 데이터 타입 변경

- 1) astype('변경할 타입')
- 2) pd.to_numeric(변경할 컬럼) : 적절한 숫자 타입으로 변경

==> 컬럼안에 nan이 있다면 numeric()사용

```python
bike3['운동량'].astype('float64') # could not convert string to float: '\\N'
```

```python
# 운동량과 탄소량 value 안에 \N 찾기

bike3.loc[bike3['운동량'] == r'\N','운동량'] = -1
bike3.loc[bike3['탄소량'] == r'\N','탄소량'] = -1
```

```python
bike3['운동량'] = bike3['운동량'].astype('float64') # could not convert string to float: '\\N'
bike3['탄소량'] = pd.to_numeric(bike3['탄소량'])
```

```python
print(bike3['운동량'].dtypes)
print(bike3['탄소량'].dtypes)
```

### 구별 거치대 개수

```python
bike1.head()
```

#### 총거치대수 = 거치대수lcd + 거치대수qr

```python
bike1['총거치대수'] = bike1['거치대수lcd'] + bike1['거치대수qr']
```

```python
bike1.head()
```

#### bike_group 데이터프레임 생성

- bike1 자치구를 그룹으로 총거치대수 구한 결과

```python
bike_group = bike1.groupby('자치구')['총거치대수'].sum().reset_index()
bike_group.head(3)
```

```python
bike_group = bike_group.set_index('자치구')
bike_group.head(2)
```

```python
import folium
import json

geo_simple = '../../data/dataset/skorea_municipalities_geo_simple.json'
geo_data = json.load(open(geo_simple, encoding='utf-8'))
geo_data
```

```python
bike_group.head()
```

```python
m = folium.Map([37.5502, 126.982], zoom_start=11, tiles='Stamen Toner')

folium.Choropleth(
    geo_data = geo_data,
    data = bike_group['총거치대수'],
    columns=['자치구','총거치대수'],
    key_on='properties.name',
    legend_name = '자치구 별 총 거치대수',
    fill_color = 'YlGnBu',
    fill_opacity = 0.6,
    line_opacity = 0.4
).add_to(m)

m
```
