```python
headers = {'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/102.0.5005.141 Whale/3.15.136.29 Safari/537.36'}
```

```python
def get_day_list(item_code, page_no):
    '''
    일자별 시세 페이지별 수집
    '''
    url = f'https://finance.naver.com/item/sise_day.naver?code={item_code}&page={page_no}'

    res = requests.get(url, headers=headers)
    
    soup = BeautifulSoup(res.text, 'lxml')

    table = soup.select('table')

    table = pd.read_html(str(table))[0].dropna()
    
    return table
    
```

```python
# 함수 테스트
item_code = '373220'
item_name = 'LG에너지솔루션'
table = get_day_list(item_code, 3)
table
```

```python
import time

page_no = 1

item_list = []

while page_no < 13:
    table = get_day_list(item_code, page_no)
    item_list.append(table)
    time.sleep(0.3)
    
    page_no += 1
```

```python
item_list
```

```python
df = pd.concat(item_list)
```

```python
df.head()
```

```python
# 종목코드, 종목명 컬럼 생성

df['종목코드'] = item_code
df['종목명'] = item_name
```

```python
# 컬럼 순서 변경

cols = ['종목코드','종목명','날짜','종가','전일비','시가','고가','저가','거래량']
df = df[cols]
```

```python
# 전체 함수로 작성

def get_item_list(item_code, item_name):    
    page_no = 1
    
    item_list = []
    
    while page_no < 13:
        table = get_day_list(item_code, page_no)
        item_list.append(table)
        time.sleep(0.3)
        
        page_no += 1
    
    df = pd.concat(item_list)
    
    df['종목코드'] = item_code
    df['종목명'] = item_name
    
    cols = ['종목코드','종목명','날짜','종가','전일비','시가','고가','저가','거래량']
    df = df[cols]
    
    return df  
```
