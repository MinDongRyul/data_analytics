```python
import dart_fss as dart

api_key = 'e5259cc84f1d93781ae76ea7a6793a64d76af962'
dart.set_api_key(api_key=api_key)
```

```python
# DART에 공시된 모든 회사 리스트 가져오기

corp_list = dart.get_corp_list()
corp_list[:10]
```

```python
len(corp_list)
```

```python
name = '삼성전자'
samsung = corp_list.find_by_corp_name(name, exactly=True)[0]
```

```python
# 재무제표 가져오기

fs = samsung.extract_fs(bgn_de='20210101')
# 엑셀저장
fs.save()
```

```python
# 여러 회사의 재무제표 가져오기

company_list = ['삼성전자','현대자동차','SK하이닉스','NAVER']

for name in company_list:
    company = corp_list.find_by_corp_name(name, exactly=True)[0]
    
    fs = company.extract_fs(bgn_de='20210101')
    fs.save()
```
