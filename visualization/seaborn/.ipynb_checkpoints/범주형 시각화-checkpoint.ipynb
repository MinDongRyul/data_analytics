```python
import seaborn as sns
sns.set(font_scale=1)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
```

```python
iris_df = sns.load_dataset('iris')
iris_df
```

## 1) boxplot

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=iris_df)
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=iris_df, orient='h')
```

## 2) swarmplot

```python
sns.swarmplot(data=iris_df)
```

```python
sns.swarmplot(data=iris_df, size=2)
```

## 3) catplot

```python
sns.catplot(data=iris_df)
```

## 4) violinplot

```python
sns.violinplot(data=iris_df)
```

## 5) stripplot

```python
sns.stripplot(data=iris_df)
```

## 6) countplot

```python
titanic_df = sns.load_dataset('titanic')
titanic_df.head(3)
```

```python
titanic_df.isnull().sum()
```

### 1) 성별 차트 - pandas

```python
titanic_df.groupby('sex')['sex'].count().plot.bar(color=['red','blue'])
```

```python
sns.countplot(data=titanic_df, x='sex')
```

```python
sns.countplot(data=titanic_df, y='sex')
```

#### 성별 인원수를 객실 등급별로 시각화

```python
titanic_df.groupby(['sex','class'])['who'].count().plot.bar(rot=0,figsize=(10,6))
```

```python
sns.countplot(data=titanic_df, x='sex', hue='class', palette='flare')
```

```python
tips_df = sns.load_dataset('tips')
tips_df.head()
```

#### 요일별 팁을 준 횟수 시각화
