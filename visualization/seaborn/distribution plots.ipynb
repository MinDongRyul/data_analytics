```python
import seaborn as sns
sns.set(font_scale=1)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
```

```python
iris_df = sns.load_dataset('iris')
iris_df.head()
```

### histplot

```python
sns.histplot(data=iris_df['sepal_length'])
```

```python
sns.histplot(data=iris_df['sepal_length'],bins=20)
```

```python
sns.histplot(data=iris_df['sepal_length'],bins=20, kde=True)
```

```python
sns.histplot(data=iris_df['sepal_length'],bins=20, kde=True, stat='density')
```

```python
plt.figure(figsize=(15,6))
sns.histplot(data=iris_df, x= 'sepal_length',bins=20, hue='species')
```

```python
plt.figure(figsize=(15,6))
sns.histplot(data=iris_df, x= 'sepal_length',
             bins=20, hue='species',kde=True, stat='density')
```

```python
sns.histplot(data=iris_df, x='sepal_width', y='sepal_length')
```

```python
sns.distplot(iris_df['sepal_length'])
```

### jointplot

```python
sns.jointplot(data=iris_df, x='sepal_width', y='sepal_length')
```

```python
sns.jointplot(data=iris_df, x='sepal_width', y='sepal_length', kind='hex')
```
