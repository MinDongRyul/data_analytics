```python
import seaborn as sns
sns.set(font_scale=1)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
```

```python
plt.style.use('seaborn')
```

```python
df = sns.load_dataset('flights')
df.head()
```

```python
df.shape
```

```python
sns.lineplot(data=df, x='year',y='passengers')
```

```python
sns.lineplot(data=df, x='year',y='passengers')
plt.xticks(df['year'], rotation=45);
```

#### 1) hue : 

```python
sns.lineplot(data=df, x='year',y='passengers', hue='month');
```

#### 2) palette

```python
sns.lineplot(data=df, x='year',y='passengers', hue='month', palette='Set2');
```

#### 3) style

```python
sns.lineplot(data=df, x='year',y='passengers', hue='month', palette='Set2', style='month');
```

#### 4) marker

```python
sns.lineplot(data=df, x='year',y='passengers', hue='month', palette='Set2', style='month', markers=True);
```

#### 5) err_style :   

```python
sns.lineplot(data=df, x='year',y='passengers',err_style='bars')
```
