### seaborn

- 범주형 변수 사용 지원 
- 복잡한 데이터 셋 구조를 간편하게 분석
- 내장된 테마로 matplot figure 스타일을 간단하게 제어
- 데이터에서 패턴을 나타내는 색상 팔레트를 선택하는 도구들
- 데이터 서브 셋을 비교하고 일변량이나 이변량 분포를 시각화하는 옵션

```python
import seaborn as sns
```

```python
sns.set(font_scale=1)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
```

```python
plt.style.use('seaborn')
```

```python
movie_title = ['크루엘라','극장판 귀멸의 칼날:무한열차편','학교 가는 길']
audience = [664308,2099131,20067]

data = {'영화제목':movie_title,'누적관객':audience}
df = pd.DataFrame(data)
df
```

```python
plt.bar(df['영화제목'],df['누적관객'],color=['blue','orange','green'])
plt.xlabel('영화제목')
plt.ylabel('누적관객')
plt.show()
```

### seaborn으로 차트 작성

```python
sns.barplot(data=df, x='영화제목', y='누적관객')
```

- 누적관객 수를 기준으로 정렬 차트 작성

```python
sns.barplot(data=df, x='영화제목', y='누적관객', order=df.sort_values('누적관객').영화제목)
```

```python
sns.barplot(data=df, x='영화제목', y='누적관객', order=df.sort_values('누적관객',ascending=False).영화제목)
```

```python
chart = sns.barplot(data=df, x='영화제목', y='누적관객', order=df.sort_values('누적관객',ascending=False).영화제목)

label_format = '{:,.0f}'

ylabels = [label_format.format(i) + '만 명' for i in chart.get_yticks() / 10000] # yticks : 0.00 x 1e6, 0.25 x 1e6, 0,50 x 1e6 ~ 2.00 x 1e6
chart.set_yticklabels(ylabels)
```

```python
chart = sns.barplot(data=df, y='영화제목', x='누적관객', order=df.sort_values('누적관객',ascending=False).영화제목)
 
label_format = '{:,.0f}' # 25.0 -> 25

xlabels = [label_format.format(i) + '만 명' for i in chart.get_xticks() / 10000] # yticks : 0.00 x 1e6, 0.25 x 1e6, 0,50 x 1e6 ~ 2.00 x 1e6
chart.set_xticklabels(xlabels)
```

```python
chart = sns.barplot(data=df, y='영화제목', x='누적관객', order=df.sort_values('누적관객',ascending=False).영화제목)
 
label_format = '{:,.0f}' # 25.0 -> 25

xlabels = [label_format.format(i) + '만 명' for i in chart.get_xticks() / 10000] # yticks : 0.00 x 1e6, 0.25 x 1e6, 0,50 x 1e6 ~ 2.00 x 1e6
chart.set_xticklabels(xlabels)

plt.xlabel('누적관객', fontsize=14)
plt.ylabel('영화제목', fontsize=14)
plt.title('영화 별 누적 관객 수',fontsize=18)
```
