### subplots()

```python
fig, axes = plt.subplots(2,2, figsize=(15,10))
# 전체 타이틀
fig.suptitle('여러 개의 그래프')

# 첫 번째 그래프 작성(국어, 막대차트)
axes[0,0].bar(student_df['이름'],student_df['국어'],label='국어점수')
axes[0,0].set_title('학생별 국어 점수')
axes[0,0].legend()
axes[0,0].set_xlabel('이름')
axes[0,0].set_ylabel('점수')
axes[0,0].grid(linestyle='--', linewidth=0.5)
axes[0,0].set_facecolor('lightyellow')

# 두 번째 그래프 작성(영어, 수학, 선 차트)
axes[0,1].plot(student_df['이름'],student_df['영어'],label='영어점수')
axes[0,1].plot(student_df['이름'],student_df['수학'],label='수학점수')
axes[0,1].legend()
axes[0,1].set_title('학생별 영어/수학 점수')
axes[0,1].set_xlabel('이름')
axes[0,1].set_ylabel('점수')

# 세 번째 그래프 작성(키, 막대차트(가로), xlabel, ylabel)
axes[1,0].barh(student_df['이름'],student_df['키'])
axes[1,0].set_title('학생별 키')
axes[1,0].set_xlabel('키')
axes[1,0].set_ylabel('이름')

# 네 번째 그래프 작성(사회, 선차트, 색상 지정)
axes[1,1].plot(student_df['이름'],student_df['사회'],color='lightgreen')
```
