## matplotlib (https://matplotlib.org/stable/index)

- 다양한 형태의 그래프를 통해서 데이터 시각화를 할 수 있는 라이브러리

### 범례 설정
- https://matplotlib.org/stable/api/legend_api.html

```python
import matplotlib.pyplot as plt
```

```python
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
```

```python
x = [1,2,3]
y = [2,4,8]
```

```python
plt.plot(x,y, label='데이터')
plt.legend()
plt.show()
```

- 두 개의 선을 하나의 차트에 그릴 때

```python
x1 = [1,2,3]
y1 = [1,2,3]

x2 = [1,2,3]
y2 = [1,4,7]
```

```python
plt.plot(x1,y1)
plt.plot(x2,y2)
plt.show()
```

```python
plt.plot(x1,y1, color='blue',label='data1')
plt.plot(x2,y2, color='red',label='data2')
plt.legend()
plt.show()
```

```python
plt.plot(x1,y1, color='blue')
plt.plot(x2,y2, color='red')
plt.legend(['data1','data2'])
plt.show()
```

- 범례 위치 지정하기 - 기본은 5

```python
plt.plot(x,y, label='데이터')
plt.legend(loc='upper right')
plt.show()
```

```python
plt.plot(x,y, label='데이터')
plt.legend(loc=(0.5,0.5))
plt.show()
```

><b>2) 범례 폰트 사이즈 지정</b>

```python
plt.plot(x1,y1, color='blue')
plt.plot(x2,y2, color='red')
plt.legend(['data1','data2'],fontsize=18,ncol=2)
plt.show()
```

><b>3) 범례 행 수 지정</b>

- [문제] : data1(x : 1, 2, 3; y : 1, 2, 3) 과 data2(x : 1, 2, 3; y : 1, 100, 200) 을 그래프로 출력하시오.

```python
plt.title('그래프', fontdict = {'size':20})
data1 = {'x':[1,2,3], 'y':[1,2,3]}
data2 = {'x':[1,2,3], 'y':[1,100,200]}

plt.plot('x','y',data=data1, color='blue')
plt.plot('x','y',data=data2, color='red')
plt.legend(['data1','dat2'],ncol=2)
plt.show()

```
