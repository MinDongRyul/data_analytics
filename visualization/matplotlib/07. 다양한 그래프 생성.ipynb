```python
import matplotlib.pyplot as plt
```

```python
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
```

### 1) 막대차트(기본)

```python
lables = ['강백호','서태웅','정대만']
values = [175,187,184]

plt.bar(labels,values)
```

><b>색상</b>

```python
plt.bar(lables,values, color='r')
```

```python
colors=['r','g','b']

plt.bar(lables,values, color=colors)
```

```python
colors=['r','g','b']

plt.bar(labels,values, color=colors, )
```

><b>축 간격 조정</b>

```python
lables = ['강백호','서태웅','정대만']
values = [175,187,184]

plt.bar(lables, values)
plt.ylim(175,190)
```

><b>막대 너비 조정</b>

```python
plt.bar(labels, values, width=0.5)
```

><b>축 레이블 rotation 설정</b>

```python
plt.bar(labels, values, width=0.3)
plt.xticks(rotation=45)
plt.yticks(rotation=45)
plt.show()
```

><b>막대 </b>

```python
xticks = ['1번학생','2번학생','3번학생']

plt.bar(labels, values)
plt.xticks(labels,xticks)
plt.show()
```

### 2) 막대차트(심화)

**수평막대**

```python
plt.barh(labels, values)
```

```python
plt.barh(labels, values)
plt.xlim(170, 195)
```

**막대 패턴**(https://matplotlib.org/stable/gallery/shapes_and_collections/hatch_style_reference.html)

```python
bar = plt.bar(labels, values)
bar[0].set_hatch('/')
bar[1].set_hatch('x')
bar[2].set_hatch('*')
```

#### 텍스트

```python
bar = plt.bar(labels, values)
plt.ylim(170,195)

# 막대 높이를 이용한 텍스트 설정
for idx, rect in enumerate(bar):
    plt.text(idx, rect.get_height() + 0.5, values[idx], ha='center')
```
