### 기본 차트 생성

><b>리스트를 그래프로 나타내기</b>

```python
plt.plot([10,20,30,40])
# plt.show()
```

```python
x = [1,2,3,4]
y = [12,43,25,15]

plt.plot(x,y)
```

><b>파이썬 딕셔너리를 그래프로 나타내기</b>

```python
dict1 = {'data_x':[1,2,3,4,5], 'data_y':[2,3,5,10,8]}

plt.plot('data_x','data_y',data=dict1)
```

### 타이틀 설정

```python
x = [1,2,3,4]
y = [12,43,25,15]

plt.title('꺽은선 차트') # 한글 설정 시 오류 발생

g = plt.plot(x,y)
```

### 한글 폰트 설정

```python
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['font.size'] = 15
```

```python
x = [1,2,3,4]
y = [12,43,25,15]

plt.title('꺽은선 차트') # 한글 설정 시 오류 발생

g = plt.plot(x,y)
```

### 마이너스 설정

```python
x = [1,2,3,4]
y = [12,-5,25,15]

plt.title('꺽은선 차트') # 한글 설정 시 오류 발생

g = plt.plot(x,y)
```

```python
plt.rcParams['axes.unicode_minus'] = False
```

```python
x = [1,2,3,4]
y = [12,-5,25,15]

plt.title('꺽은선 차트') # 한글 설정 시 오류 발생

g = plt.plot(x,y)
```

### 그리드

**전체설정**

```python
plt.plot(x,y)
plt.grid(True)
```

**x축설정**

```python
plt.plot(x,y)
plt.grid(True, axis='x')
```

**y축설정**

```python
plt.plot(x,y)
plt.grid(True, axis='y')
```
